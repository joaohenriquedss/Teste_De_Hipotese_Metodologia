# Hipótese : A falta de conhecimento sobre IoT tem uma relação significativa com pessoas que não estejam na área de TI.

# Os dados analisados já passaram por um pré processamentos de dados.
## O formulário foi constituído por 5 questões 
### 3 acertos ou menos cai na categoria : Falta de conhecimento
### 4 ou mais cai na categoria : Tem conhecimento


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format='svg'
```

###  Coluna : TI
* Sim = É da área de TI
* Não = Não é da área de TI

### Coluna : Conhecimento
* 0 = Falta de conhecimento em IoT
* 1 = Tem conhecimento em IoT

```python
data = pd.read_csv('iot.csv')
data.head(10)
```

# Quantidade de entrevistados

```python
print('Participantes : ' + str(data.TI.count()))
```

# Quem é da área de TI ou não.

```python
# Com Pandas
data.TI.value_counts().plot(kind='pie', autopct='%.2f%%')
plt.axis('equal')  # Para deixar o gráfico redondo
plt.ylabel('Trabalham com TI ?')
fig = plt.figure(figsize=(10,5))
plt.show()
```

# Conhecimento sobre IoT

```python
# Com Pandas
data.CONHECIMENTO.value_counts().plot(kind='pie', autopct='%.2f%%')
plt.axis('equal')  # Para deixar o gráfico redondo
plt.ylabel('CONHECIMENTO IoT?')
fig= plt.figure(figsize=(10,5))
plt.show()
```

# Preparando dados

```python
data_ti_0 = data[data['TI'] == 'Sim']
data_ti_0 = data_ti_0[data_ti_0['CONHECIMENTO'] == 0]

data_ti_1 = data[data['TI'] == 'Sim']
data_ti_1 = data_ti_1[data_ti_1['CONHECIMENTO'] == 1]
```

```python
data_n_ti_0 = data[data['TI'] == 'Não']
data_n_ti_0 = data_n_ti_0[data_n_ti_0['CONHECIMENTO'] == 0]

data_n_ti_1 = data[data['TI'] == 'Não']
data_n_ti_1 = data_n_ti_1[data_n_ti_1['CONHECIMENTO'] == 1]
```

# Entrevistados sem conhecimento em IoT

```python
print('Entrevistados sem conhecimento em IoT\n')
print('É da área de TI :', str(data_ti_0.TI.count()))
print('Não é da área de TI :', str(data_n_ti_0.TI.count()))
```

```python
grupos = ['TI: Sim', 'TI: Não']
valores = [(data_ti_0.TI.count()), (data_n_ti_0.TI.count())]
plt.bar(grupos, valores)
plt.title('Sem conhecimento sobre IoT')
fig= plt.figure(figsize=(10,5))
plt.show()
```

```python
# Com Pandas
a = [data_ti_0, data_n_ti_0]
result_n = pd.concat(a)
result_n.TI.value_counts().plot(kind='pie', autopct='%.2f%%')
plt.axis('equal')  # Para deixar o gráfico redondo
plt.ylabel('Sem conhecimento sobre IoT')
fig= plt.figure(figsize=(10,5))
plt.show()
```

# Entrevistados com conhecimento em IoT

```python
print('Entrevistados com conhecimento em IoT\n')
print('É da área de TI :', str(data_ti_1.TI.count()))
print('Não é da área de TI :', str(data_n_ti_1.TI.count()))
```

```python
grupos = ['TI: Sim', 'TI: Não']
valores = [(data_ti_1.TI.count()), (data_n_ti_1.TI.count())]
plt.bar(grupos, valores)
plt.title('Com conhecimento sobre IoT')
fig= plt.figure(figsize=(10,5))
plt.show()
```

```python
# Com Pandas
a = [data_ti_1, data_n_ti_1]
result_n = pd.concat(a)
result_n.TI.value_counts().plot(kind='pie', autopct='%.2f%%')
plt.axis('equal')  # Para deixar o gráfico redondo
plt.ylabel('Com conhecimento sobre IoT')
fig= plt.figure(figsize=(10,5))
plt.show()
```

# Foi possível observar com essa pequena quantidade de dados que a hipótese inicial é verdadeira mas para uma análise mais profunda seria necessário uma coleta de dados maior.
